```python
import matplotlib.pyplot as plt

# 결합확률질량함수 데이터
joint_pmf_data = {
    (1, 1): 0.1,
    (1, 2): 0.2,
    (1, 3): 0.2,
    (2, 1): 0.3,
    (2, 2): 0.05,
    (2, 3): 0.15
}

# X와 Y의 주변확률밀도함수를 계산
f_X = {}
f_Y = {}

# X의 주변확률밀도함수 계산
for x, y in joint_pmf_data.keys():
    f_X[x] = f_X.get(x, 0) + joint_pmf_data[(x, y)]

# Y의 주변확률밀도함수 계산
for x, y in joint_pmf_data.keys():
    f_Y[y] = f_Y.get(y, 0) + joint_pmf_data[(x, y)]

# Y축의 범위를 0부터 1까지, X축의 범위를 동일하게 설정하여 그래프를 그리기
plt.figure(figsize=(12, 5))

# Y축의 범위 설정
y_min = 0
y_max = 1

# X축의 범위 설정
x_min = 0.5
x_max = 3.5

# X의 주변확률질량함수 시각화
plt.subplot(1, 2, 1)
plt.bar(f_X.keys(), f_X.values(), color='skyblue')
plt.title('X의 주변확률질량함수 f_X(x)')
plt.xlabel('X')
plt.ylabel('확률')
plt.ylim(y_min, y_max)
plt.xlim(x_min, x_max)

# Y의 주변확률질량함수 시각화
plt.subplot(1, 2, 2)
plt.bar(f_Y.keys(), f_Y.values(), color='lightcoral')
plt.title('Y의 주변확률질량함수 f_Y(y)')
plt.xlabel('Y')
plt.ylabel('확률')
plt.ylim(y_min, y_max)
plt.xlim(x_min, x_max)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 예시 데이터 (이미 계산되어 있다고 가정)
f_X = {1: 0.5, 2: 0.5}
f_Y = {1: 0.4, 2: 0.25, 3: 0.35}

# 범위 설정
y_min = 0
y_max = 1
x_min = 0.5
x_max = 3.5

# X의 주변확률질량함수 그래프
plt.subplot(1, 2, 1)
plt.plot(list(f_X.keys()), list(f_X.values()), 'bo')  # 점
plt.vlines(list(f_X.keys()), y_min, list(f_X.values()), colors='blue')  # 수선의 발
plt.title('Marginal PMF of X')
plt.xlabel('X value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max) + 1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# Y의 주변확률질량함수 그래프
plt.subplot(1, 2, 2)
plt.plot(list(f_Y.keys()), list(f_Y.values()), 'ro')  # 점
plt.vlines(list(f_Y.keys()), y_min, list(f_Y.values()), colors='red')  # 수선의 발
plt.title('Marginal PMF of Y')
plt.xlabel('Y value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max) + 1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.tight_layout()
plt.show()

```

```python
import numpy as np

# 국어 성적과 수학 성적 데이터
korean_scores = [75, 65, 50, 70, 80]
math_scores = [65, 70, 70, 55, 60]

# 데이터를 NumPy 배열로 변환 (2행 5열)
data = np.array([korean_scores, math_scores])

# 공분산 계산
# bias=True → N으로 나눔 (모집단 공분산)
# bias=False → N-1로 나눔 (표본 공분산, 기본값)
covariance = np.cov(data, bias=True)[0, 1]

# 결과 출력
print("Covariance:", covariance)
```

```python
import numpy as np

# 데이터
x = np.array([4, 6, -4, 10])
y = np.array([-1, 5, 3, 18])

# 상관계수 계산
correlation = np.corrcoef(x, y)[0, 1]

# 결과 출력
print("Correlation Coefficient:", correlation)
```
